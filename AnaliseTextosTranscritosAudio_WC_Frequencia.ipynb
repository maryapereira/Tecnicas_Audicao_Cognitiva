```python
texto = 'Motivo do atraso geral de. Os cota sem o motivos atrasam todos. Ninguém disse é a gente pode negociar o pagamento dessa parcela que tem aberto eu consigo um desconto para dieta de dias sem. É tamém de saudade se não consegue efetuar antes do senhor não consegue. Seis de. Erra descansam a verificar o valor já tá sempre sem lançam uma mente. O valor atualizado com juros e multa até dia seis estaria novecentos e quarenta e um e setenta e três com juros e com desconto novecentos reais e sessenta centavos. Oito seis a secretaria que manipula boletos por email s m s. Hoje o seu prefere que uma de boleto por bem ou por s m s. De lá sai nesse mesmo. Pode nos trazer porque no meio de um ano. A eu tenho e veio da camila que. Que ao os. Se o pagamento direto na boca do caixa amenta. Bem no caixa. Thais mendes já garante a manequim diários. Chat na.'
```

```python
textoSplit = texto.split()
```

```python
import nltk
nltk.download('punkt')

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

stop_words = set(stopwords.words('portuguese')) 
stop_words.add(('senhor','saudade','ta','de', 'veio'))

  
word_tokens = word_tokenize(texto) 
  
texto_sem_stopwods = [w.lower() for w in word_tokens if not w in stop_words] 
  
texto_sem_stopwods = [] 
  
for w in word_tokens: 
    if w not in stop_words: 
        texto_sem_stopwods.append(w) 
  
print(word_tokens) 
print(texto_sem_stopwods)
```

```python
textoSplit
```

```python
def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=['dtype', 'Name', 'object', 'texto', 'o', 'tá', 'saudade', 'De'],
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=70
    ).generate(str(data))
    
    fig = plt.figure(1, figsize=(16, 12))
    plt.axis('off')
    plt.imshow(wordcloud)
    plt.show()
```

```python
show_wordcloud(texto_sem_stopwods)
```

```python
frequencia_texto = nltk.FreqDist(texto_sem_stopwods)

df_frequencia = pd.DataFrame({"Palavra": list(frequencia_texto.keys()),
                              "Frequência": list(frequencia_texto.values())})

df_frequencia.nlargest(columns="Frequência", n=30).set_index('Palavra').plot(kind='barh', color='r', figsize=(16, 8))
plt.gca().invert_yaxis()
plt.title('Frequência das palavras do texto (removendo as stopwords)')
plt.show()
```
